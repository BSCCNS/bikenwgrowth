```python
import folium
import pandas as pd
from IPython.display import display

#read coordinates of the lines that will be drawn. (I am using the attribute lines from the edges file, is it correct?)
def read_coordinates(file_path):
    df = pd.read_csv(file_path)
    coordinates = []
    for row in df.itertuples():
        line = row.geometry.split('LINESTRING (')[1].split(')')[0].split(', ')
        line = [i.split() for i in line]
        line = [[float(i[1]), float(i[0])] for i in line]
        coordinates.append(line)
    return coordinates

bike_coordinates = read_coordinates('copenhagen_biketrack_simplified_edges.csv')
car_coordinates = read_coordinates('copenhagen_carall_simplified_edges.csv')

################################To add a satellite map##############################################
#token='pk.eyJ1IjoidG9jdW5oYSIsImEiOiJja2lxMWpwNG0wOWJvMnhsMzR5cHg2NHo1In0.2UhDfgs_Cf5c4T9yJRmvPQ'

#tileurl = 'https://api.mapbox.com/v4/mapbox.satellite/{z}/{x}/{y}@2x.png?access_token=' + str(token)

#m = folium.Map(
#    location=[55.6761, 12.5683], zoom_start=12, tiles=tileurl, attr='Mapbox')
####################################################################################################
m = folium.Map(location=[55.6761, 12.5683],
               zoom_start=12)

#Draw and add bike network layer
my_PolyLine=folium.PolyLine(locations=bike_coordinates,weight=1.5, color = 'blue')
my_PolyLine.add_to(folium.FeatureGroup(name='Bike Network').add_to(m))

#Draw and add car network layer
my_PolyLine2=folium.PolyLine(locations=car_coordinates,weight=1.5, color = 'red')
my_PolyLine2.add_to(folium.FeatureGroup(name='Car Network').add_to(m))


folium.LayerControl().add_to(m)
display(m)
```
